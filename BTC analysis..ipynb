```python
import pandas as pd


df = pd.read_csv(r"C:\Users\stefa\Downloads\BTC1.csv", delimiter=';')

# Visualizza il DataFrame
df.info()
```

```python
df = pd.DataFrame(df)

# Conversione della colonna 'Data' in tipo datetime
df['Data'] = pd.to_datetime(df['Data'], format='%d.%m.%Y')

df
```

```python
import pandas as pd


df['volatility'] = df['Massimo'] - df['Minimo']
df['volatilita_percentuale'] = (df['Massimo'] - df['Minimo']) / df['Apertura'] * 100


df
```

```python
import pandas as pd
df['rendimento'] = df['Ultimo'].pct_change() * 100
df
```

```python
df.info()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
sns.relplot(x = 'Data', y = 'volatility', kind = 'line', data = df)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Creazione del grafico
plt.figure(figsize=(10, 6))
plt.plot(df['Data'], df['rendimento'], color='red')
plt.title('Rendimento YoY')
plt.xlabel('Data')
plt.ylabel('rendimento')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df['Data'], df['volatility'], color='blue')
plt.title('Volatility YoY')
plt.xlabel('Data')
plt.ylabel('volatility')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Creazione del grafico
plt.figure(figsize=(10, 6))
plt.plot(df['Data'], df['volatilita_percentuale'], color='green')
plt.title('Volatilità percentuale YoY')
plt.xlabel('Data')
plt.ylabel('volatilita_percentuale')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
f, ax = plt.subplots(figsize = (6,6))
cmap = sns.cubehelix_palette(as_cmap = True, dark = 0 ,light = 1, reverse = True)
sns.kdeplot(x = 'Data', y = 'volatility', data = df, cmap = cmap, n_levels = 100, shade = True)
plt.show()
```
