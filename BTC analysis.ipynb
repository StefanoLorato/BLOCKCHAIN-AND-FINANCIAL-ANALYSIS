```python
import pandas as pd


df = pd.read_csv(r"C:\Users\stefa\Downloads\BTC1.csv", delimiter=';')

# Visualizza il DataFrame
df.info()
```

```python
df = pd.DataFrame(df)

# Conversione della colonna 'Data' in tipo datetime
df['Data'] = pd.to_datetime(df['Data'], format='%d.%m.%Y')

df.info()
```

```python
import pandas as pd


# Calcoliamo la volatilità come differenza tra massimo e minimo (High-Low)
df['volatility'] = df['Massimo'] - df['Minimo']
df['volatilita_percentuale'] = (df['Massimo'] - df['Minimo']) / df['Apertura'] * 100


df
```

```python
import pandas as pd

# Supponiamo che 'df' sia il tuo DataFrame con i dati OHLC
# 'open', 'high', 'low', 'close' rappresentano rispettivamente i prezzi di apertura, massimo, minimo e chiusura

# Calcola il rendimento giornaliero come variazione percentuale tra il prezzo di chiusura di un giorno e il prezzo di apertura del giorno successivo
df['rendimento'] = df['Ultimo'].pct_change() * 100

# Puoi quindi fare riferimento a df['rendimento'] per il rendimento calcolato

df
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
sns.relplot(x = 'Data', y = 'volatility', kind = 'line', data = df)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Creazione del grafico
plt.figure(figsize=(10, 6))
plt.plot(df['Data'], df['rendimento'], color='red')
plt.title('Rendimento  nel tempo')
plt.xlabel('Data')
plt.ylabel('rendimento')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Creazione del grafico
plt.figure(figsize=(10, 6))
plt.plot(df['Data'], df['volatilita_percentuale'], color='green')
plt.title('Volatilità percentuale nel tempo')
plt.xlabel('Data')
plt.ylabel('volatilita_percentuale')
plt.grid(True)
plt.show()
```
